# Exploratory Data Analysis on Model Performance
This notebook provides an analysis of various model performance metrics based on the provided dataset. The dataset includes information on frame processing times, processing FPS, model accuracy, and video characteristics.

Each section of the notebook presents visualizations and insights to help understand the performance of different models and formats under varying conditions.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the CSV file
file_path = './report.csv'
data = pd.read_csv(file_path)

# Data Cleaning and Preprocessing
data['Average Frame Processing Time (sec)'] = data['Average Frame Processing Time'].str.replace(' sec', '').astype(float)
data['Processing FPS'] = data['Processing FPS'].str.replace(' fps', '').astype(float)
data['Target Class Score (%)'] = data['Target Class Score'].str.replace('%', '').astype(float)
data['Total time (sec)'] = data['Total time'].str.replace(' sec', '').astype(float)
data['Video Duration (sec)'] = data['Video Duration'].str.replace(' sec', '').astype(float)
data['Video FPS'] = data['Video FPS'].str.replace(' fps', '').astype(float)
data.head()
```

## Processing Time Trends by Model Format and Model
This section analyzes the trends in average frame processing time across different model formats and models.

```python
plt.figure(figsize=(14, 7))
sns.lineplot(data=data, x='Model_Format', y='Average Frame Processing Time (sec)', hue='Model_Name', marker='o')
plt.title('Trends in Average Frame Processing Time by Model Format and Model Name')
plt.xlabel('Model Format')
plt.ylabel('Average Frame Processing Time (seconds)')
plt.legend(title='Model Name', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Accuracy Trends by Model Format
This plot shows how target class accuracy varies across different model formats for each model.

```python
plt.figure(figsize=(14, 7))
sns.lineplot(data=data, x='Model_Format', y='Target Class Score (%)', hue='Model_Name', marker='o')
plt.title('Trends in Target Class Accuracy by Model Format and Model Name')
plt.xlabel('Model Format')
plt.ylabel('Target Class Accuracy (%)')
plt.legend(title='Model Name', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Side-by-Side Comparisons of All Model Formats for Target Class Accuracy
This bar chart provides a side-by-side comparison of target class accuracy for each model, grouped by model format.

```python
plt.figure(figsize=(14, 8))
sns.barplot(data=data, x='Model_Name', y='Target Class Score (%)', hue='Model_Format')
plt.title('Side-by-Side Comparison of Target Class Accuracy by Model Format')
plt.xlabel('Model Name')
plt.ylabel('Target Class Accuracy (%)')
plt.legend(title='Model Format', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Comparison of Video Duration and Accuracy
This scatter plot examines the relationship between video duration and target class accuracy across models.

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='Video Duration (sec)', y='Target Class Score (%)', hue='Model_Name')
plt.title('Comparison of Video Duration and Target Class Accuracy')
plt.xlabel('Video Duration (seconds)')
plt.ylabel('Target Class Accuracy (%)')
plt.legend(title='Model Name', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Analysis of Processing FPS Across Different Models
This bar chart compares processing FPS across different models and formats.

```python
plt.figure(figsize=(12, 7))
sns.barplot(data=data, x='Model_Name', y='Processing FPS', hue='Model_Format')
plt.title('Comparison of Processing FPS Across Different Models')
plt.xlabel('Model Name')
plt.ylabel('Processing FPS')
plt.legend(title='Model Format', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```
